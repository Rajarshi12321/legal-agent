```python
class DocumentRetriever:
    def __init__(self, index_path, embeddings, k=8, allow_dangerous_deserialization=True):
        """
        Initialize the FAISS retriever by loading the local FAISS index.
        
        :param index_path: Path to the FAISS index.
        :param embeddings: Embeddings instance to be used with the index.
        :param k: Number of nearest neighbors to retrieve.
        :param allow_dangerous_deserialization: Flag for enabling dangerous deserialization.
        """
        # Load the FAISS index from the given path and create a retriever with search settings.
        self.retriever = (
            FAISS.load_local(
                index_path,
                embeddings,
                allow_dangerous_deserialization=allow_dangerous_deserialization
            )
            .as_retriever(search_kwargs={"k": k})
        )
    
    def retrieve(self, query):
        """
        Retrieve documents relevant to the provided query.
        
        :param query: The search query string.
        :return: Retrieved documents.
        """
        docs = self.retriever.invoke(query)
        return docs

    def pretty_print_docs(self, docs):
        """
        Pretty print the retrieved documents.
        
        :param docs: List of documents to print.
        """
        for i, doc in enumerate(docs, start=1):
            print(f"Document {i}:")
            print(doc)
            print("-" * 40)

# # Example usage:
# if __name__ == "__main__":
#     # Assume `embeddings` is defined and available for use
#     # For instance: embeddings = SomeEmbeddingModel(...)
    
#     # Initialize the retriever with the index path and embeddings
#     retriever_instance = DocumentRetriever("../faiss_index", embeddings, k=8)

#     # Define your query string
#     query = "Your query here"
    
#     # Retrieve documents and print them
#     documents = retriever_instance.retrieve(query)
#     retriever_instance.pretty_print_docs(documents)
```
